# Partícula em uma caixa unidimensional

$$-\frac{\hbar^2}{2m}\frac{d^2}{dx^2}\psi(x) + V\psi(x) = E\psi(x) \tag{1}$$

Dentro da caixa o potencial é zero, logo:

$$-\frac{\hbar^2}{2m}\frac{d^2}{dx^2}\psi(x) = E\psi(x) \tag{2}$$

Vamos adotar uma possível solução para a função de onda:

$$\psi(x) = A\sin(sx) +B\cos(rx)$$

Derivando duas vezes:

$$\frac{d}{dx}\psi(x) = sA\cos(sx) - rB\sin(rx)$$

$$\frac{d^2}{dx^2}\psi(x) = - s^2A\sin(sx) - r^2B\cos(rx)$$

Usando a derivada segunda e a solução adotada para a função de onda na **Equação 2**, temos que:

$$-\left(- s^2A\sin(sx) - r^2B\cos(rx)\right) = \frac{2mE}{\hbar^2}\left(A\sin(sx) +B\cos(rx)\right)$$

Mudando o sinal no lado esquerdo da equação

$$\left( s^2A\sin(sx) + r^2B\cos(rx)\right) = \frac{2mE}{\hbar^2}\left(A\sin(sx) +B\cos(rx)\right)$$

Agora, distribuindo o fator $\frac{2m}{\hbar^2}$

$$s^2A\sin(sx) + r^2B\cos(rx) = \frac{2mE}{\hbar^2}A\sin(sx) + \frac{2mE}{\hbar^2}B\cos(rx)$$

Observando a equação, notamos que:

$$ s^2 = r^2 = \frac{2mE}{\hbar^2} $$

Logo:

$$ s = r = \frac{\sqrt{2mE}}{\hbar} $$

Portanto, a solução proposta para a função de onda ,$\psi(x)$, se torna:

$$\psi(x) = A\sin\left(\frac{(2mE)^{1/2}}{\hbar}x\right) +B\cos\left(\frac{(2mE)^{1/2}}{\hbar}x\right)$$

Aplicando as condições de contorno para $\psi(0)$, vemos:

$$\psi(0) = A\sin\left(\frac{(2mE)^{1/2}}{\hbar}0\right) +B\cos\left(\frac{(2mE)^{1/2}}{\hbar}0\right)$$

$$0 = A\sin\left(0\right) +B\cos\left(0\right)$$

$$ B = 0$$

Obs.: $\sin(0) = 0$ e $\cos(0) = 1$

Deste modo, obtemos que B = 0 e vamos aplicar a condição de contorno para $\psi(L)$: 

$$\psi(L) = A\sin\left(\frac{(2mE)^{1/2}}{\hbar}L\right)$$

Obs.: Como B = 0 o outro termo da equação desaparece

$$0 = A\sin\left(\frac{(2mE)^{1/2}}{\hbar}L\right)$$

Visto que o valor de **A** não pode ser zero, o seno precisa ser zero para que a equação seja verdadeira. Portanto:

$$\sin\left(\frac{(2mE)^{1/2}}{\hbar}L\right) = 0$$

Para que o seno seja zero, ele precisa ser um multiplo inteiro de $\pi$, com base no circulo trigonometrico, então

$$\frac{(2mE)^{1/2}}{\hbar}L =  n\pi$$

$$\frac{(2mE)^{1/2}}{\hbar} =  \frac{n\pi}{L}$$

Por fim a equação da função de onda se torna

$$\boxed{\psi(x) = A\sin\left(\frac{n\pi}{L}x\right)} \tag{3}$$

Vamos determinar a energia usando a função de onda encontrada. Vamos derivar a **Equação (3)** duas vezes

$$\frac{d}{dx}\psi(x) = \frac{n\pi}{L}A\cos\left(\frac{n\pi}{L}x\right)$$

$$\frac{d^2}{dx^2}\psi(x) = -\frac{n^2\pi^2}{L^2}A\sin\left(\frac{n\pi}{L}x\right)$$

Voltando a **Equação (2)**, obtemos

$$-\frac{\hbar^2}{2m}\left(-\frac{n^2\pi^2}{L^2}A\sin\left(\frac{n\pi}{L}x\right)\right) = EA\sin\left(\frac{n\pi}{L}x\right)$$

Fazendo as simplificações temos

$$\boxed{E = \frac{n^2\pi^2\hbar^2}{2mL^2}}$$

Agora queremos encontrar o valor da constante de normalização, para isso usaremos a seguinte expressão:

$$\int_{-\infty}^{\infty} |\psi|^2dx = 1$$

Usando a equação acima, integrando no intervalo de 0 até L e aplicando a função de onda encontrada no item **(a)**, obtemos:

$$\int_{0}^{L} A^2\sin^2\left(\frac{n\pi}{L}x\right)dx = 1$$

Desenvolvendo a integral e sabendo que $\sin^2\left(\frac{n\pi}{L}x\right) = \frac{1}{2}(1 - \cos(\frac{2n\pi}{L}x))$ :

$$\frac{A^2}{2}\left[\int_{0}^{L}dx - \int_{0}^{L}\cos(\frac{2n\pi}{L}x)\right] = 1$$

$$\frac{A^2}{2}\left[L - 0 - (\sin(\frac{2n\pi L}{L}) - \sin(\frac{2n\pi 0}{L}))\right] = 1$$

O termos que contem seno tem valor zero, uma vez que $\sin(0) =0 $ e $\sin(2n\pi) = 0$. Logo:

$$\frac{A^2L}{2} = 1$$

Por fim, o valor de **A** é:

$$\boxed{A = \left(\frac{2}{L} \right)^{1/2}}$$

Deste modo, encontramos a constante de normalização e podemos agora aplica-la na **Equação 3**:

$$\boxed{\psi(x) = \left(\frac{2}{L} \right)^{1/2}\sin\left(\frac{n\pi}{L}x\right)} \tag{4}$$

Observamos então que a função de onda vária seguindo um padrão, e que vária de acordo com o número quantico principal **n**.
